## a) Which one of these Bayesian networks are correct representations of the preceding information?

    Each telescope has an independent Fi. N and Fi are inputs for the telescope and Mi is the output. The telescope's N and Fi also influence our Mi. N conditionally depends on Fi and Mi. P(N|Mi, Fi) cannot be simplified to P(N|Mi) so graph (i) does not accurately represent the information. However graphs (ii) and (iii) are correct representations of the network.

## d) Suppose M1 = 1 and M2 = 3. What are the possible number of stars if we assume no prior constraints on the value of N?

    With no prior constraints on the value of N and these values for Mi, the possible number of stars that we could potentially observe is 2, 4, and any value 6 or greater.
    

## e) What is the most likely number of stars given these observations?

    Since we know that N can only be of values 2, 4 and 6+, we just need to evaluate these values.
    
    The probability that each telescope under or over counted the starts when N = 2 is
    e^2 * (1-f)^2 / 4
    
    The probability when N = 4 is
    e * (1-f) * f / 4
    
    The probability when N = 6 is
    f^2
    
    Becuase e is always greater than f, we can assume that e*(1-f)/4 > f. Therefore the probability for N=2 is greater than that for N in {4, 6, ...}

## a) Can knowledge of the value of projector_plugged_in affect your belief in the value of Sam_reading_book?

    No, the value of projector_plugged_in does not affect belief in the value of sam_reading_book. Power_in_wire will check the dependency flow from projector_plugged_in to sam_reading_book.


## b) Can knowledge of screen_lit_up affect your belief in Sam_reading_book?

    Yes, the knowledge of screen_lit_up does affect belief in the value of sam_reading_book. The path from screen_lit_up to sam_reading_book is not checked by any value, so they are dependent.


## c) Can knowledge of the value of projector_plugged_in affect your belief in the value of Sam_reading_book given that you have observed a value for a screen_lit_up?

    Yes, knowledge of the value projector_plugged_in given that I have observed a value for screen_lit_up does affect belief in the value of sam_reading_book. Our path from (b) necessitates checks for dependency because power_in_wire exists within that path. This means that power_in_wire will not be checking the dependency flow.


## d) Which variables could have their probabilities changed if just lamp_works was observed?

    The variables projector_lamp_on, screen_lit_up, and ray_says_"screen_is_dark" will have their probabilities changed if just lamp_works is observed. The path from these variables to lamp_words has no variables that check for dependency flows.


## e) Which variables could have their probabilities changed if just power_in_projector was observed?

    Only input variables light_switch_on, ray_is_awake, lamp_works, and  mirror_working are NOT affected by observing power_in_projector. This is because the path from power_in_projector to every other variable has some variable that checks dependency flows.
